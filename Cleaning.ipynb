```python
import os
```

```python
all_mrs = os.listdir("data")
```

```python
only_lumbar = list(filter(lambda x: x.find("_L") > -1, all_mrs))
```

```python
import csv
```

```python
mri_lookup = {}
with open('sheet1.csv') as mri_stats:
    csv_reader = csv.reader(mri_stats, delimiter=',')
    for idx,row in enumerate(csv_reader):
        r = row[0].split(',')
        if (idx > 0):
            if r[0] in mri_lookup:
                if r[2] in mri_lookup[r[0]]:
                    mri_lookup[r[0]][r[2]] += 1
                else:
                    mri_lookup[r[0]][r[2]] = 1
            else: 
                mri_lookup[r[0]] = {}
```

```python
mri_max_slice = {}
for mri in mri_lookup:
    if (len(mri_lookup[mri].keys()) > 0):
        mri_max_slice[mri] = max(mri_lookup[mri], key=mri_lookup[mri].get)
```

```python
mri_max_slice_L = {}

for mri in mri_max_slice:
    if (mri.find('_L') > -1 and int(mri_max_slice[mri]) > 0):
        mri_max_slice_L[mri] = int(mri_max_slice[mri])
```

```python
s1 = set(mri_max_slice_L.keys())
```

```python
s2 = set(only_lumbar)
```

```python
cleaned_mri_list = {}
file_size = 0
for m in mri_max_slice_L:
    if os.path.exists('data/' + m):
        for f in os.listdir('data/' + m):
            m_slice = mri_max_slice_L[m]
            if ('0000' + str(m_slice)) in f:
                cleaned_mri_list[m] = f
                file_size += os.path.getsize('data/' + m + '/' + f)
                
```
