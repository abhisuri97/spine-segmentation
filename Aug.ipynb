```python
import csv
```

```python
d = {}
with open('Spine_Measurements.csv') as csvfile:
    csv_reader = csv.reader(csvfile, delimiter=',')
    for idx, row in enumerate(csv_reader):
        if (idx > 0):
            if (row[0].find('_L')>-1):
                if row[0] in d:
                    if (row[2] in d[row[0]]):
                        d[row[0]][row[2]] += 1
                    else:
                        d[row[0]][row[2]] = 1
                else:
                    d[row[0]] = {row[2]: 1}

for r in d.keys():
    s = list(d[r].keys())
    d[r] = int(sorted(s, reverse=True, key=lambda k: d[r][k])[0])
    
print(d)
```
