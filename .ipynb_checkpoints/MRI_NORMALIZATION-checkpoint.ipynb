```python
import pydicom
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
ds = pydicom.dcmread('mri_L/MR_0004_L/00008_1.3.12.2.1107.5.2.31.30664.2013101612353327089557754.dcm')
```

```python
flat = np.array(ds.pixel_array).flatten()
```

```python
m = np.mean(flat)
```

```python
s = np.std(flat)
```

```python
cutoff = m + 0.5*s
r = np.clip(ds.pixel_array + 0, 0, cutoff)
f = np.interp(r, (r.min(), r.max()), (0, 255))
f = np.where(abs(f - 255) < 50, 0, f)
f = np.interp(f, (f.min(), f.max()), (0, 255))
fig = plt.figure()

plt.subplot(1, 2, 1)
plt.imshow(f, cmap="bone", vmin=0, vmax=255)

plt.subplot(1, 3, 3)
plt.hist(f.flatten(), 100)

plt.show()

```

```python
import os
total = 0
images = []
for idx, x in enumerate(os.listdir('mri_L')):
    for y in os.listdir('mri_L/{}'.format(x)):
        if (y.find('dcm') > -1):
            x = np.array(pydicom.dcmread('mri_L/{}/{}'.format(x, y)).pixel_array)
            images.append(np.interp(x, (x.min(), x.max()), (0, 255)))
```

```python
def show_images(images):
    n = len(images)
    f = plt.figure(figsize=(20,10))
    for i in range(n):
        f.add_subplot(27, 37, i + 1)
        plt.imshow(images[i])
    plt.show(block=True)
    return f
```

```python
f = show_images(images)
```

```python
f.savefig('test.png')
```

```python
m = np.mean([x[0] for x in [[np.mean(x.flatten()), np.std(x.flatten())] for x in images]])
```

```python
s = np.mean([x[1] for x in [[np.mean(x.flatten()), np.std(x.flatten())] for x in images]])
```

```python
import cv2
```

```python
scale = 56
rescaled = []
for i in images:
    rescaled.append(cv2.resize(i, dsize=(scale, scale), interpolation=cv2.INTER_CUBIC))
fl = [np.mean([y.flatten()[x] for y in rescaled]) for x in range(scale*scale)]
plt.imshow(np.array(fl).reshape((scale,scale)), cmap='bone')
```

```python
fixed_imgs = []
for x in images:
    fixed_imgs.append(np.where(x - (m + 2*s) > 0, 0, x))
```

```python
def mean_image(scale, images):
    rescaled = []
    for i in images:
        rescaled.append(cv2.resize(i, dsize=(scale, scale), interpolation=cv2.INTER_CUBIC))
    fl = [np.mean([y.flatten()[x] for y in rescaled]) for x in range(scale*scale)]
    plt.imshow(np.array(fl).reshape((scale,scale)), cmap='bone')
```
