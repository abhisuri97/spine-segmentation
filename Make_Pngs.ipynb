# Make PNG images from MRI dicoms

```python
import pydicom
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
def conv_png(dcm_path, file_name, dest_dir='images'):
    print(dcm_path)
    ds = pydicom.dcmread(dcm_path)
    shape = ds.pixel_array.shape
    img_2d = ds.pixel_array.astype(float)
    img_2d *= 255.0/img_2d.max()
    img_2d_scaled = np.clip(np.uint(img_2d) * 1.5, 0, 255)
    im = Image.fromarray(img_2d_scaled.astype(np.uint8))
    im.save(dest_dir + '/' + file_name + '.png')
    print('saved ' + dcm_path)
```

```python
import os
```

```python
for idx, x in enumerate(os.listdir('mri_L')):
    for dcm in os.listdir('mri_L/' + x):
        if (dcm.find('.dcm') > -1):
            conv_png('mri_L/{}/{}'.format(x,dcm), x)
```
